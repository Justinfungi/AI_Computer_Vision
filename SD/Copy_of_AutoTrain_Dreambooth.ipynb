```python
from diffusers import DiffusionPipeline, StableDiffusionXLImg2ImgPipeline
import torch

prj_path = "Justinfungi/SDXL"
model = "stabilityai/stable-diffusion-xl-base-1.0"

# Normal pipe
pipe_normal = DiffusionPipeline.from_pretrained(
    model,
    torch_dtype=torch.float16,
)
pipe_normal.to("cuda")

# Lora pipe
pipe = DiffusionPipeline.from_pretrained(
    model,
    torch_dtype=torch.float16,
)
pipe.to("cuda")
pipe.load_lora_weights(prj_path, weight_name="pytorch_lora_weights.safetensors")

# refiner
refiner = StableDiffusionXLImg2ImgPipeline.from_pretrained(
    "stabilityai/stable-diffusion-xl-refiner-1.0",
    torch_dtype=torch.float16,
)
refiner.to("cuda")

# Run

prompts = [
    "A photo of a cat, in the style of National Geographic",
    "An illustration of a dragon, in the style of Harry Potter",
    "A 3D render of a car, in the style of Pixar",
    "A painting of a landscape, in the style of Bob Ross",
    "A photorealistic image of a flower, in the style of David Attenborough",
    "A cartoon image of a robot, in the style of The Jetsons",
    "An abstract image of a face, in the style of Pablo Picasso",
    "A surreal image of a melting clock, in the style of Salvador Dali",
    "A futuristic image of a city, in the style of Blade Runner",
    "A dark fantasy image of a knight, in the style of Game of Thrones"
]
prompts = [i + " in laser cut" + "black and white" for i in prompts]
for prompt in prompts:
  #prompt = "A cute dog in laser cut"
  seed = 42
  generator = torch.Generator("cuda").manual_seed(seed)
  image = pipe(prompt=prompt, generator=generator).images[0]
  image = refiner(prompt=prompt, generator=generator, image=image).images[0]
  image.save(f"generated_image_{count}.png")

  image = pipe(prompt=prompt, generator=generator).images[0]
  image = refiner(prompt=prompt, generator=generator, image=image).images[0]
  image.save(f"generated_image_{count}_normal.png")
  count+=1
```

```python
import os
import zipfile

zipf = zipfile.ZipFile('output.zip', 'w', zipfile.ZIP_DEFLATED)

for root, dirs, files in os.walk('.'):
    for file in files:
        if file.endswith('.png'):
            zipf.write(file)

zipf.close()
```
